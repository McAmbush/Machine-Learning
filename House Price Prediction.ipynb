```python
features = ['sqft_living','sqft_above','grade','bathrooms','sqft_living15']
```

```python
X = df[features]
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sn
```

```python
Y = df[['price']]
```

```python
plt.figure(figsize = (10,5))
plt.scatter(X['sqft_living'],Y)
```

```python
sn.regplot(X['sqft_living'],Y['price'])
```

```python
X.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
from sklearn.metrics import r2_score,mean_squared_error
```

```python
scaler = StandardScaler()
```

```python
x = scaler.fit_transform(X)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
train_x,test_x,train_y,test_y = train_test_split(x,Y,test_size = 0.2)
```

```python
lr.fit(train_x,train_y)
```

```python
lr_pred = lr.predict(test_x)
```

```python
print(r2_score(lr_pred,test_y))
```

```python
print(mean_squared_error(lr_pred,test_y))
```

```python
poly = PolynomialFeatures(degree = 2)
```

```python
x_pl = poly.fit_transform(train_x)
```

```python
lr2 = LinearRegression()
```

```python
lr2.fit(x_pl,train_y)
```

```python
test_x_pl = poly.fit_transform(test_x)
```

```python
pl_pred = lr2.predict(test_x_pl)
```

```python
print(r2_score(test_y,pl_pred))
```

```python
print(mean_squared_error(test_y,pl_pred))
```
